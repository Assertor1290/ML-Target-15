## Difference between randn and randint

```python
# np.random.rand is for Uniform distribution (in the half-open interval [0.0, 1.0))
# np.random.randn is for Standard Normal (aka. Gaussian) distribution (mean 0 and variance 1)

# NORMAL DISTRIBUTION: is a probability distribution where probability of x is highest at centre and lowest in the ends
# UNIFORM DISTRIBUTION: probability of x is constant.
```

```python
# Return random integers from `low` (inclusive) to `high` (exclusive).

# Return random integers from the "discrete uniform" distribution of
# the specified dtype in the "half-open" interval [`low`, `high`). If
# `high` is None (the default), then results are from [0, `low`).
# x = np.random.randint?
```

```python
# Return a sample (or samples) from the "standard normal" distribution.
# If positive int_like arguments are provided, `randn` generates an array
# of shape ``(d0, d1, ..., dn)``, filled
# with random floats sampled from a univariate "normal" (Gaussian)
# distribution of mean 0 and variance 1.
# x = np.random.randn?
```

```python
x = np.random.randn(1000)
```

```python
# out = plt.hist?
```

```python
out = plt.hist(x,bins=100)
```

```python
x = np.random.randint(-4,4,1000)
```

```python
out = plt.hist(x,bins=120)
```

```python
# x : Make a pie chart of array *x*.  The fractional area of each wedge is
# given by ``x/sum(x)``.
# explode : If not *None*, is a ``len(x)`` array which specifies the fraction
# of the radius with which to offset each wedge.
out = plt.pie([10,20,30])
```

```python
out = plt.pie([10,20,30], explode=[0,0.1,0])
```

```python
out = plt.pie([10,20,30], explode=[0,0.1,0],shadow=True,radius=0.5)
```

## Seaborn

```python
import seaborn as sns
```

```python
x = np.random.randn(1000)
```

```python
plt.figure(figsize=(4, 4))
sns.distplot(x, bins=30)
```

```python
plt.figure(figsize=(4, 4))
sns.distplot(x, bins=30,kde=False)
```

```python
plt.figure(figsize=(4, 4))
sns.distplot(x, bins=30,kde=False,rug=True)
```
