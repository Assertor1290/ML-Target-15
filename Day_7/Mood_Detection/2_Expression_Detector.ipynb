```python
import cv2
import dlib
import numpy as np
```

```python
# beneficial than open cv as can detect tilted face, partial face
detector = dlib.get_frontal_face_detector()
```

```python
# for landmark like open cv had haarcascade
predictor = dlib.shape_predictor('../datasets/shape_predictor_68_face_landmarks.dat')
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()
    # using gray instead of frame for computation part to avoid lag
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = detector(gray)
    for face in faces:
        landmarks = predictor(gray,face)
        
        # .y taking vertical coordinates
        lip_up = landmarks.parts()[62].y
        lip_down = landmarks.parts()[66].y
        
        if lip_down-lip_up > 5:
            print("Mouth Open")
        else:
            print("Mouth Closed")
            
        # save data as we saved in face recognition    
        # expression = np.array([[point.x, point.y] for point in landmarks.parts()[:48]])
        # print(expression)
        # print(expression.flatten())
        
        # problem with above is we dont want it to in a particular area without structure of anything    
        expression = np.array([[point.x - face.left() , point.y-face.top()] for point in landmarks.parts()[48:]])
        print(expression.flatten())
        
        # only show mouth coordinates 
        for point in landmarks.parts()[48:]:
            cv2.circle(frame,(point.x,point.y),1,(255,0,0),2)
    
    if ret:    
        cv2.imshow("My Screen", frame)
       
    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
