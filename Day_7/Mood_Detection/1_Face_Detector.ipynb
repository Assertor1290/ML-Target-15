```python
import cv2
import dlib
```

```python
# beneficial than open cv as can detect tilted face, partial face
detector = dlib.get_frontal_face_detector()
```

```python
# for landmark like open cv had haarcascade
predictor = dlib.shape_predictor('../datasets/shape_predictor_68_face_landmarks.dat')
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()
    # using gray instead of frame for computation part to avoid lag
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = detector(gray)
    for face in faces:
        landmarks = predictor(gray,face)
        # This shows the points for all parts of face
        # print(landmarks.parts())
        
        # index starts from 0
        # nose = landmarks.parts()[27]
        # print(nose.x,nose.y)
        
        # image, center, radius, color(BGR), thickness
        # cv2.circle(frame,(nose.x,nose.y),2,(255,0,0),3)
        for point in landmarks.parts():
            cv2.circle(frame,(point.x,point.y),1,(255,0,0),2)
    
    # print rectangle coordinates of face
    # print(faces)
    
    if ret:    
        cv2.imshow("My Screen", frame)
       
    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
