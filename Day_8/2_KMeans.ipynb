```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Here we are generating data for our KMeans model
# here n_smaples is default 100
X, y = make_blobs(centers=3, random_state=42)
# we can ignore y but we will have to accept it
# X, _ = make_blobs(centers=3, random_state=42)
```

```python
# has two features
X
```

```python
import seaborn as sns
```

```python
# all of same color
# sns.scatterplot(X[:,0],X[:,1])
# this provides color or classes
# sns.scatterplot(X[:,0],X[:,1],hue=y)
```

```python
# our KMeans data, we dont have clusters, simply data
sns.scatterplot(X[:,0],X[:,1])
```

```python
# Kmeans says we have to minimize loss
# i.e. minimize distance from corresponding cluster center
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(X)
```

```python
# this is the y generated for each point in X
model.labels_
```

```python
y_gen =model.labels_
```

```python
sns.scatterplot(X[:,0],X[:,1],hue=y_gen)
```

```python
model.cluster_centers_
```

```python
sns.scatterplot(X[:,0],X[:,1],hue=y_gen)
for center in model.cluster_centers_:
    plt.scatter(center[0],center[1], s=60)
```
