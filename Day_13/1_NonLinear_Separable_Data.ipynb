## Creating new dimension

```python
def dim(X):
    X1 = X[:,0]
    X2 = X[:,1]
    # X3 is value for third dimension, can do anything
    X3 = X1**2 + X2**2
    # Previously each example had 2 features X1 and X2; (in 2D)
    # Now, we want each exammple to have 3 features, X1, X2, X3; (in 3d)
    # No of features = no of dimensions
    
    # new X ; no of rows remain same.; no of column = no of features =3
    # zeros(shape, dtype=float, order='C')
    # Return a new array of given shape and type, filled with zeros.
    X_ = np.zeros((X.shape[0], 3))
    # all rows and all columns except last column 
    X_[:,:-1] = X
    # all rows and last column
    X_[:,-1] = X3
    
    return X_
```

```python
X_ = dim(X)
```

#### Plot 3d data

```python
# Part of matplotlib
from mpl_toolkits.mplot3d import Axes3D 
```

```python
def plot3d(X, show=True):
    # The purpose of using plt.figure() is to create a figure object.
    # The whole figure is regarded as the figure object. It is necessary to explicitly use plt.figure() 
    # when we want to tweak the size of the figure and when we want to add multiple Axes objects in a single figure.
    fig = plt.figure(figsize=(10,10))
    # An Axes3D object is created just like any other axes using the projection=‘3d’ keyword.
    # The default mode for matplotlib is to produce 2D plots. The option projection 
    # becomes available after import Axes3D from mpl_toolkits as the base _axes object will be added.
    # https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html#getting-started
    ax = fig.add_subplot(111,projection='3d')
    X1 = X[:,0]
    X2 = X[:,1]
    X3 = X[:,2]
    
    # (Position of data points: X1,X2,X3, s=size of points)
    ax.scatter(X1,X2,X3,zdir='z',s=20,c=y,depthshade=True)
    if(show==True):
        plt.show()
    return ax
```

```python
ax = plot3d(X_)
```

#### Visualize Decision Boundary

```python
wts = model.coef_
bias = model.intercept_
```

```python
wts
```

```python
bias
```

```python
# https://stackoverflow.com/questions/36013063/what-is-the-purpose-of-meshgrid-in-python-numpy
xx,yy = np.meshgrid(range(-1,2),range(-1,2))
print(xx)
print(yy)
```

```python
z = -(wts[0,0]*xx + wts[0,1]*yy+bias)/wts[0,2]
print(z)
```

```python
ax = plot3d(X_, show=False)
ax.plot_surface(xx,yy,z,alpha=0.2)
plt.show()
```
