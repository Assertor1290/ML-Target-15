```python
#PIL is Python Imaging library. Image is a module
from PIL import Image
#to depict pixels in form of array
import numpy as np
#to plot image. matplotlib.pyplot is a collection of 
#command style functions that make Matplotlib work like MATLAB. Each pyplot function makes some change to a figure
import matplotlib.pyplot as plt
```

```python
#open an image
pil=Image.open("goku.jpg")
```

```python
pil
```

```python
#As every image is made of pixels, we can depict it in the form of an array
arr=np.array(pil)
#arr diplays points
```

```python
arr
```

```python
arr.shape
```

```python
#imshow:-Display data as an image, i.e., on a 2D regular raster.
plt.imshow(arr)
#OR plt.imshow(pil)
```

```python
cut=arr[:5,:5]
plt.imshow(cut)
```

```python
cutface=arr[50:125,50:125]
plt.imshow(cutface)
```

```python
temp=arr.copy()
tempblack=arr.copy()
```

```python
temp[:,:,[1,2]]=0
plt.imshow(temp)
```

```python
temp
```

```python
tempblack[:,:,:]=0
plt.imshow(tempblack)
```

```python
under=np.arange(60).reshape(4,5,3)
under
```

```python
undertemp=under.copy()
undertemp[:,:,[1,2]]=0
undertemp
```

```python
grayscale=arr.mean(axis=2) #across depth
plt.imshow(grayscale)
```

```python
#If we do not use cmap, the image will show a multicolour grayscale picture. cmap basically controls the colour map.
#A color mapping may be referred to as the algorithm that results in the mapping function or the algorithm that transforms the image colors
plt.imshow(grayscale,cmap="gray")
```

```python
plt.imshow(grayscale,cmap='hot')
```

```python
plt.imshow(grayscale,cmap='jet')
```
